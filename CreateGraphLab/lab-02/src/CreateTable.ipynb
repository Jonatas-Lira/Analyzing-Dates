```python
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
datas = pd.read_csv('cars.csv')

price_freq = datas['price_usd'].value_counts()

freq_byPrice = price_freq.reset_index()

print(tabulate(freq_byPrice, headers=["Preço", "Frequência"], tablefmt="pretty"))
```

```python
plt.figure(figsize=(12, 6))
plt.hist(datas['price_usd'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribuição de Preços em USD')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
