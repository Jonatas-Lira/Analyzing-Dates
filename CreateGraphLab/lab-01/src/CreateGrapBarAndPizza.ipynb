```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datas = pd.read_csv('cars.csv')
color_counts = datas['color'].value_counts()

colors = {
    'black': 'white',
    'silver': 'silver',
    'blue': 'blue',
    'white': 'black',
    'grey': 'grey',
    'red': 'red',
    'green': 'green',
    'lightgray': 'lightgray',
    'brown': 'brown',
    'violet': 'violet',
    'yellow': 'yellow',
    'orange': 'orange'
}
```

```python
# Gr√°fico de barras
plt.figure(figsize=(10, 6))
bars = color_counts.plot(kind='bar', color=[colors.get(color, 'lightgray') for color in color_counts.index])

# Contornos das barras
for bar in bars.patches:
    bar.set_edgecolor('black')
    bar.set_linewidth(1)

plt.title('Contagem de Carros por Cor')
plt.xlabel('Cores')
plt.ylabel('Quantidade de Carros')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
patches, texts, autotexts = plt.pie(color_counts.values, labels=color_counts.index, colors=colors, autopct='%1.1f%%')
plt.title("Gr√°fico de pizza")

plt.legend(patches, [f'{label} - {percent:.1f}%' for label, percent in zip(color_counts.index, color_counts/sum(color_counts)*100)], title="Cores", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Distribui√ß√£o de Carros por Cor')
plt.tight_layout()
plt.show()
```
